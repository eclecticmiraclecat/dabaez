## formatting

```python
name = 'IBM'
shares = 100
price = 32.2
```

```python
print(name, shares, price)
```

```python
print(f'{name :10s} {shares :10d} {price :10.2f}')
```

#### :10s
    
    10 = width
    s = string
    d = digit
    f = floating point

```python
print(f'{name :>10s} {shares :>10d} {price :>10.2f}')
```

#### :>10s

    > = right justified
